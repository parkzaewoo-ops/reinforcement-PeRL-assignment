```python
import numpy as np
import pandas as pd

import os
import json
import dspy
from datasets import load_dataset
import sqlite3
import matplotlib.pyplot as plt

# Test connection to the configured local LLM (Ollama on http://localhost:11434)
print("Using dspy version:", getattr(dspy, '__version__', 'unknown'))
try:
    # Instantiate LM with required 'model' argument to avoid __init__() missing error
    lm = dspy.LM("openai/Qwen/Qwen3-Next-80B-A3B-Instruct", api_base="http://211.47.56.81:7972/v1", api_key="token-abc123")
    # Optionally configure dspy's global LM reference
    dspy.configure(lm=lm)
    # send a short test prompt
    resp = lm("Say 'hello' in one short sentence.")
    print('LM responded:', resp)
except Exception as e:
    print('Connection failed:', e)

# Disable pyarrow for pandas conversion to avoid ArrowDtype compatibility issues
# This forces the datasets library to use standard pandas dtypes instead of Arrow types
os.environ["DATASETS_DISABLE_PYARROW"] = "1"

try:
    # Load the BIRD SQL dataset
    # split parameter specifies which dataset split to load
    dataset = load_dataset(
        "birdsql/bird_sql_dev_20251106",
        split="dev_20251106"
    )
    print(f"✓ Successfully loaded dataset with {len(dataset)} examples")

except Exception as e:
    # Catch and display any errors during loading
    # Common errors: JSON parse errors, ArrowDtype issues, network errors
    print(f"Error: {e}")

# Filter dataset to get only "challenging" difficulty examples
challenging_data = dataset.filter(lambda x: x['difficulty'] == 'challenging')
print(f"✓ Found {len(challenging_data)} challenging examples")

# Split into train (80%) and test (20%)
split_dataset = challenging_data.train_test_split(test_size=0.2, seed=42)

# Assign to train and test variables
train_data = split_dataset['train']
test_data = split_dataset['test']

print(f"✓ Data split complete:")
print(f"  - Train data: {len(train_data)} examples ({len(train_data)/len(challenging_data)*100:.1f}%)")
print(f"  - Test data: {len(test_data)} examples ({len(test_data)/len(challenging_data)*100:.1f}%)")
print(f"  - Total: {len(challenging_data)} examples")

# If you want them as lists instead of Dataset objects:
train_list = list(train_data)
test_list = list(test_data)

# Define DSPy signature for text-to-SQL task
class TextToSql(dspy.Signature):
    """
    You are a Text-to-SQL model.

    Given a database schema and a natural language question,
    generate a single valid SQLite SQL query that answers the question.

    - Use only the tables and columns that appear in the given schema.
    - Add appropriate WHERE conditions, JOINs, and GROUP BY / ORDER BY
      when necessary to answer the question.
    - Do not return any explanation or comments. Return only the final SQL query.
    """

    schema: str = dspy.InputField(
        desc="Database schema description, including table names and columns."
    )
    question: str = dspy.InputField(
        desc="Natural language question that must be answered using the schema."
    )
    sql: str = dspy.OutputField(
        desc="A single SQLite SQL query that correctly answers the question. No explanation."
    )

# Create ChainOfThought predictor for SQL generation
generate_sql = dspy.ChainOfThought(TextToSql)

# Function to create DSPy examples with fixed schema_info
def create_dspy_examples_with_fixed_schema(raw_data, schema_text: str):
    examples = []
    for item in raw_data:
        db_id = item.get('db_id', 'unknown')

        ex = dspy.Example(
            question=item.get('question', ''),
            schema=schema_text,
            sql=item.get('SQL', ''),
            db_id=db_id
        ).with_inputs('question', 'schema')

        examples.append(ex)

    return examples

# Create DSPy examples with fixed schema_info
schema_text = schema_info.strip()

train_examples = create_dspy_examples_with_fixed_schema(train_list, schema_text)
test_examples  = create_dspy_examples_with_fixed_schema(test_list, schema_text)

print(f"✓ Created {len(train_examples)} train examples and {len(test_examples)} test examples with fixed schema_info.")
```

```python
def normalize_sql(sql: str) -> str:
    if sql is None:
        return ""
    s = sql.strip()

    if s.endswith(";"):
        s = s[:-1]

    s = " ".join(s.split())

    return s.lower()

def predict_sql_no_db(example, predictor=generate_sql, verbose=False):
    pred = predictor(
        schema=example.schema,
        question=example.question
    )

    pred_sql = getattr(pred, "sql", "") or ""
    gold_sql = example.sql or ""

    if verbose:
        print("=== Example ===")
        print(f"DB ID      : {example.db_id}")
        print(f"Question   : {example.question}")
        print(f"Gold SQL   : {gold_sql}")
        print(f"Pred SQL   : {pred_sql}")
        print("================\n")

    return pred_sql, gold_sql

def evaluate_exact_match_no_db(predictor, examples, max_examples=None, verbose_every=0):
    if max_examples is not None:
        eval_examples = examples[:max_examples]
    else:
        eval_examples = examples

    n_total = len(eval_examples)
    n_correct = 0

    for idx, ex in enumerate(eval_examples, start=1):
        pred_sql, gold_sql = predict_sql_no_db(ex, predictor=predictor, verbose=False)

        norm_pred = normalize_sql(pred_sql)
        norm_gold = normalize_sql(gold_sql)

        if norm_pred == norm_gold:
            n_correct += 1

        if verbose_every and idx % verbose_every == 0:
            print(f"[eval] {idx}/{n_total} processed "
                  f"(current EM: {n_correct/idx*100:.2f}%)")

    em = n_correct / n_total * 100 if n_total > 0 else 0.0
    print(f"Final Text-based Exact Match: {em:.2f}% ({n_correct}/{n_total})")

    return em, n_correct, n_total

def text_sql_em_metric(example, pred, trace=None) -> float:
    gold_sql = normalize_sql(example.sql)
    pred_sql = normalize_sql(getattr(pred, "sql", ""))

    return 1.0 if gold_sql == pred_sql else 0.0
```

```python
print(">>> Baseline evaluation (no DB, string-based EM) on a small subset")
em_small, correct_small, total_small = evaluate_exact_match_no_db(
    predictor=generate_sql,
    examples=test_examples,
    max_examples=20,
    verbose_every=5
)
```

```python
tele = dspy.BootstrapFewShot(
    metric=text_sql_em_metric,
    max_bootstrapped_demos=4,
    max_labeled_demos=32
)

print(">>> Compiling generate_sql with BootstrapFewShot ...")
compiled_generate_sql = tele.compile(
    generate_sql,
    trainset=train_examples
)
print(">>> Done: compiled_generate_sql is ready.")
```

```python
print(">>> Evaluation after BootstrapFewShot optimization")
em_bs, correct_bs, total_bs = evaluate_exact_match_no_db(
    predictor=compiled_generate_sql,
    examples=test_examples,
    max_examples=47,
    verbose_every=10
)

print(f"[BootstrapFewShot] Final Text-based EM: {em_bs:.2f}% "
      f"({correct_bs}/{total_bs})")
```

```python
def show_bootstrapfewshot_demos_from_tele(tele, max_print=3):
    candidate_attrs = ["demos", "labeled_demos", "bootstrapped_demos", "trainset"]
    demos = None
    chosen_attr = None

    for attr in candidate_attrs:
        if hasattr(tele, attr):
            maybe = getattr(tele, attr)
            try:
                _ = len(maybe)
                demos = maybe
                chosen_attr = attr
                break
            except TypeError:
                continue

    if demos is None:
        print("No suitable demos-like attribute found on teleprompter.")
        print("Available attrs:", [a for a in dir(tele) if not a.startswith("_")])
        return

    print(f"Using tele.{chosen_attr} as demos (len={len(demos)})")
    print("-" * 80)

    for i, ex in enumerate(demos[:max_print], start=1):
        db_id = getattr(ex, "db_id", "N/A")
        question = getattr(ex, "question", "")
        sql = getattr(ex, "sql", "")
        schema = getattr(ex, "schema", "")

        print(f"[Demo {i}] db_id: {db_id}")
        print(f"Question:\n{question}\n")
        print(f"Gold SQL:\n{sql}\n")
        schema_preview = str(schema)[:400].replace("\n", " ")
        print(f"Schema snippet (first 400 chars):\n{schema_preview}...\n")
        print("-" * 80)


print(">>> Inspecting demos used by BootstrapFewShot (from tele)...")
show_bootstrapfewshot_demos_from_tele(tele, max_print=3)
```

```python
from typing import Any

def safe_sql_str(x: Any) -> str:
    if x is None:
        return ""
    if x is ... or isinstance(x, type(...)):
        return ""
    if isinstance(x, str):
        return x
    try:
        return str(x)
    except Exception:
        return ""

def normalize_sql(sql: Any) -> str:
    s = safe_sql_str(sql)
    s = s.strip()

    if s.endswith(";"):
        s = s[:-1]

    s = " ".join(s.split())

    return s.lower()

def sql_token_similarity(gold_sql: Any, pred_sql: Any) -> float:
    gold = normalize_sql(gold_sql)
    pred = normalize_sql(pred_sql)

    gold_tokens = set(gold.replace(",", " ").split())
    pred_tokens = set(pred.replace(",", " ").split())

    if len(gold_tokens) == 0:
        return 0.0

    return len(gold_tokens & pred_tokens) / len(gold_tokens)

def text_sql_em_metric(example, pred, trace=None) -> float:
    gold_sql = normalize_sql(example.sql)
    pred_sql = normalize_sql(getattr(pred, "sql", ""))

    return 1.0 if gold_sql == pred_sql else 0.0
```

```python
baseline_results = []

for ex in test_examples:
    pred_sql, gold_sql = predict_sql_no_db(
        ex, predictor=generate_sql, verbose=False
    )
    baseline_results.append((pred_sql, gold_sql))

bs_results = []

for ex in test_examples:
    pred_sql, gold_sql = predict_sql_no_db(
        ex, predictor=compiled_generate_sql, verbose=False
    )
    bs_results.append((pred_sql, gold_sql))

print("✓ baseline_results collected:", len(baseline_results))
print("✓ bs_results collected:", len(bs_results))

baseline_sims = [sql_token_similarity(g, p) for p, g in baseline_results]
bs_sims       = [sql_token_similarity(g, p) for p, g in bs_results]
```
